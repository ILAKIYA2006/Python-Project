```python
from transformers import pipeline
import pandas as pd
import matplotlib.pyplot as plt
classifier = pipeline("sentiment-analysis")
texts = [
    "I love the new phone I bought!",
    "The delivery was late and the package was damaged.",
    "It’s okay, not too great but not bad either.",
    "Worst experience ever, very disappointed.",
    "Absolutely amazing service and product!"
]
results = classifier(texts)

for text, result in zip(texts, results):
    print(f"Text: {text}")
    print(f"Sentiment: {result['label']}, Score: {result['score']:.2f}")
    print("-" * 50)
labels = [res['label'] for res in results]
df = pd.DataFrame(labels, columns=['Sentiment'])
counts = df['Sentiment'].value_counts()

plt.pie(counts, labels=counts.index, autopct='%1.1f%%', startangle=140)
plt.title("Sentiment Distribution")
plt.show()
```
